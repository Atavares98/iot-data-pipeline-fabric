```python
# CÉLULA 1 — Silver: Dados Meteorológicos

from pyspark.sql.functions import to_timestamp, col
from pyspark.sql.types import DoubleType

# Ler Bronze
df_weather_bronze = spark.read.format("parquet").load("Files/bronze/weather_raw")

# Renomear colunas
df_weather_silver = df_weather_bronze \
    .withColumnRenamed("time", "event_timestamp") \
    .withColumnRenamed("temperature_2m", "temperature_ext") \
    .withColumnRenamed("relative_humidity_2m", "humidity_ext")

# Converter tipos
df_weather_silver = df_weather_silver \
    .withColumn("event_timestamp", to_timestamp(col("event_timestamp"))) \
    .withColumn("temperature_ext", col("temperature_ext").cast(DoubleType())) \
    .withColumn("humidity_ext", col("humidity_ext").cast(DoubleType()))

# Remover nulos
df_weather_silver = df_weather_silver.dropna()

# Guardar
df_weather_silver.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("silver_weather")
print(f"Silver weather: {df_weather_silver.count()} registos")
```

```python
# CÉLULA 2 — Silver: Dados Meteorológicos Forecast

from pyspark.sql.functions import to_timestamp, col
from pyspark.sql.types import DoubleType

# Ler Bronze forecast
df_weather_bronze_forecast = spark.read.format("delta").load("Files/bronze/weather_forecast")

# Renomear colunas
df_weather_silver_forecast = df_weather_bronze_forecast \
    .withColumnRenamed("time", "event_timestamp") \
    .withColumnRenamed("temperature_2m", "temperature_ext") \
    .withColumnRenamed("relative_humidity_2m", "humidity_ext")

# Converter tipos
df_weather_silver_forecast = df_weather_silver_forecast \
    .withColumn("event_timestamp", to_timestamp(col("event_timestamp"))) \
    .withColumn("temperature_ext", col("temperature_ext").cast(DoubleType())) \
    .withColumn("humidity_ext", col("humidity_ext").cast(DoubleType()))

# Remover nulos
df_weather_silver_forecast = df_weather_silver_forecast.dropna()

# Guardar
df_weather_silver_forecast.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("silver_weather_forecast")
print(f"Silver weather forecast: {df_weather_silver_forecast.count()} registos")
```
